# 1. (Problem 2.6)

```python
def Qform(a, b, c):
    x_1 = (-b + cmath.sqrt(b ** 2 - 4 * a * c)) / (2 * a)
    x_2 = (-b - cmath.sqrt(b ** 2 - 4 * a * c)) / (2 * a)
    x_plus, x_minus = np.array([x_1, x_2])
    return x_plus, x_minus
```

```python
a = 1
b = 10**8
c = 1

x_plus, x_minus = Qform(a, b, c)

print("A)", x_minus)
print("B)", x_plus)

x_plus2 = c /(a * x_minus)

print("C)", x_plus2)
```

# 5. (Problem 3.25)


```python
def f(x):
    return 2 + 5 * np.sin(x) + 0.1 * np.sin(30 * x)
```

```python
p = np.linspace(0, 2 * np.pi, 128)
fp = f(p)
```

```python
def g(x):
    return 2 + 5 * np.sin(x)

gp = g(p)
```

```python
plt.figure()
plt.plot(p, fp, label = "f(x)")
plt.plot(p, gp, label = "g(x)")
plt.legend()

plt.show()
```

```python
def fprime(x):
    return 5 * np.cos(x) + 3 * np.cos(30 * x)

fprimep = fprime(p)

def gprime(x):
    return 5 * np.cos(x)

gprimep = gprime(p)

fprimepfwd = fwd_diff(p, fp)
```

```python
plt.figure()
plt.plot(p, fprimep, label = "f'(x)")
plt.plot(p, gprimep, label = "g'(x)")
plt.plot(fprimepfwd[0], fprimepfwd[1], label = "f'(x) (forward diff)")
plt.legend()

plt.show()
```

```python
def fwd_diff2(x,y):
    h=x[2]-x[0]
    yprime = np.array([])
    for i in range(len(x)-2):
        yprime=np.append(yprime,(y[i+2]-y[i])/h)
    xdiff=x[:-2]
    return xdiff, yprime 

fprimepfwd2 = fwd_diff2(p, fp)
```

```python
plt.figure()
plt.plot(p, fprimep, label = "f'(x)")
plt.plot(p, gprimep, label = "g'(x)")
plt.plot(fprimepfwd2[0], fprimepfwd2[1], label = "f'(x) (forward diff)")
plt.legend()

plt.show()
```

```python
def smooth(x,n):
    A=np.zeros((n,len(x)))
    for i in range(0,n):
        A[i]=np.roll(x,i)
    sdata=np.mean(A,axis=0)
    sdata=np.roll(sdata,np.int32(-n/2))
    return sdata

smoothn1 = smooth(fprimepfwd[1], 21)
```

```python
plt.figure()
plt.plot(p, gprimep, label = "g'(x)")
plt.plot(fprimepfwd2[0], smoothn1[:-1], label = "f'(x) (forward diff)")
plt.legend()

plt.show()
```
