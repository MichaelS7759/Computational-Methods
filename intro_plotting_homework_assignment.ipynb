```python
import numpy as np, matplotlib.pyplot as plt
```

# Problem 1

```python
x = np.linspace(0, 2 * np.pi, 100)
x
```

```python
def f(x):
    return np.exp(6 * np.sin(x))

f = f(x)
```

```python
plt.scatter(x, f, 25, c="red", marker="d")
plt.title("Problem 1 A")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.show()
```

```python
plt.plot(x, f, c="green")
plt.title("Problem 1 B")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.show()
```

```python
df = pd.read_csv('NoisyFunction.dat', header=None, sep='\s+').to_numpy().T
t, y = df
```

# Problem 4

```python
iris = pd.read_csv('iris.csv')
iris
```

```python
pl = iris["Petal Length (cm)"].to_numpy()
plbin_min = np.floor(np.min(pl))
plbin_max = np.ceil(np.max(pl)) + 1
plbins = np.arange(plbin_min, plbin_max, 1)
print(plbins)

sl = iris["Sepal Length (cm)"].to_numpy()
slbin_min = np.floor(np.min(sl))
slbin_max = np.ceil(np.max(sl)) + 1
slbins = np.arange(slbin_min, slbin_max, 1)
print(slbins)
```

```python
fig, ax = plt.subplots(2, 1, figsize = (9, 11))
fig.suptitle("Problem 4", fontsize= 64)

ax[0].set_title("Petal Length Frequency")
ax[0].hist(pl, plbins)
ax[0].set_xlabel("Petal Length (cm)")
ax[0].set_ylabel("# Flowers")

ax[1].set_title("Sepal Length Frequency")
ax[1].hist(sl, slbins)
ax[1].set_xlabel("Sepal Length (cm)")
ax[1].set_ylabel("# Flowers")

fig.show()
```
